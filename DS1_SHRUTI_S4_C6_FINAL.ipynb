##### SEPERATING CATEGORICAL AND NUMERICAL COLUMNS:

```python
def sep_data_type(df):
    CAT=[]
    NUM=[]
    for i in df.columns:
        if df[i].nunique() <20:
            CAT.append(i)
        else:
            NUM.append(i)
    return CAT,NUM
CAT,NUM= sep_data_type(df)
table=[CAT,NUM]
print(tabulate({"Categorical":CAT,
                "Numerical": NUM}, headers = ["categorical", "Numerical"]))
            
```

##### User defined functions for level 1 analysis of categorical columns


```python
def info_of_cat(col): 
    print("Unique values are:", df[col].unique())                        #unique values
    print("Mode is: ",df[col].mode()[0])                               #mode
    print("Number of missing values are :",df[col].isnull().sum())     #missing values 
    if df[col].isnull().sum() > 0:
        print(f"\nThere are null values in the {col} column")
```

##### User defined functions for level 1 analysis of numerical columns


```python
def info_of_num(col): 
    print("Mean is: ",df[col].mean())                               #mode
    print("Number of missing values are :",df[col].isnull().sum())     #missing values 
    if df[col].isnull().sum() > 0:
        print(f"\nThere are null values in the {col} column")
```

##### MISSING VALUE TREATMENT:

```python
info_of_cat("OCCUPATION_TYPE")
```

```python
occu_tm = df.OCCUPATION_TYPE.mode()[0]
print(occu_tm)
df["OCCUPATION_TYPE"].fillna(occu_tm, inplace = True)
```

```python
info_of_cat("NAME_TYPE_SUITE")
```

```python
name_tm = df.NAME_TYPE_SUITE.mode()[0]
print(name_tm)
df["NAME_TYPE_SUITE"].fillna(name_tm, inplace = True)
```

```python
info_of_num("AMT_GOODS_PRICE")
```

```python
amt_tm = df.AMT_GOODS_PRICE.mean()
print(amt_tm)
df["AMT_GOODS_PRICE"].fillna(amt_tm, inplace = True)
```
