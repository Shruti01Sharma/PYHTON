##### SEPERATING CATEGORICAL AND NUMERICAL COLUMNS:

```python
def sep_data_type(df1):
    CAT=[]
    NUM=[]
    for i in df1.columns:
        if df1[i].nunique() <10:
            CAT.append(i)
        else:
            NUM.append(i)
    return CAT,NUM
CAT,NUM= sep_data_type(df1)
table=[CAT,NUM]
print(tabulate({"Categorical":CAT,
                "Numerical": NUM}, headers = ["categorical", "Numerical"]))
            
```

##### User defined functions for level 1 analysis of categorical columns

```python
def info_of_cat(col): 
    print("Unique values are:", df1[col].unique())                        #unique values
    print("Mode is: ",df1[col].mode()[0])                               #mode
    print("Number of missing values are :",df1[col].isnull().sum())     #missing values 
    if df1[col].isnull().sum() > 0:
        print(f"\nThere are null values in the {col} column")
    
```

##### User defined functions for level 1 analysis of numerical columns

```python
def info_of_num(col): 
    print("Unique values are:", df1[col].unique())                        #unique values
    print("Mean is: ",df1[col].mean())                               #mode
    print("Number of missing values are :",df1[col].isnull().sum())     #missing values 
    if df1[col].isnull().sum() > 0:
        print(f"\nThere are null values in the {col} column")
    
```
